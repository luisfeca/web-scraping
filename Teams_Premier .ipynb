```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Archivo excel
premier=pd.read_excel('teams.xlsx')
```

```python
premier
```

```python
premier.columns
```

```python
premier.shape
```

```python
premier.isna().sum()
```

```python
 #Ordenar los equipos por la cantidad de victorias en orden descendente
win= premier.sort_values('wins', ascending=False)

win
```

```python
# Generar una lista de colores aleatorios para cada #team_name
num_bars = len(win)
colors = np.random.rand(num_bars, 3)

# Crear el gráfico de barras con colores aleatorios
plt.bar(win['team_name'], win['wins'], color=colors)
plt.xlabel('Equipo')
plt.ylabel('Cantidad de victorias')
plt.title('Cantidad de victorias por equipo')
plt.xticks(rotation=90)

# Mostrar el gráfico de barras
plt.show()
```

```python
# Ordenar los equipos por la cantidad de victorias en casa en orden descendente
premier_sorted = premier.sort_values('wins_home', ascending=False)
premier_sorted
```

```python
# Generar una lista de colores aleatorios
num_bars = len(premier_sorted)
colors = np.random.rand(num_bars, 3)

# Crear el gráfico de barras con colores aleatorios
plt.bar(premier_sorted['team_name'], premier_sorted['wins_home'], color=colors)
plt.xlabel('Equipo')
plt.ylabel('Cantidad de victorias en casa')
plt.title('Cantidad de victorias en casa por equipo')
plt.xticks(rotation=90)

# Mostrar el gráfico de barras
plt.show()
```

```python
# Obtener los datos de goles concedidos en los minutos 61 al 70 por equipo
goles_concedidos = premier['goals_conceded_min_61_to_70']

# Ordenar los datos por goles concedidos de forma descendente
equipos_ordenados = goles_concedidos.sort_values(ascending=False)
# Generar una lista de colores aleatorios
num_bars = len(premier_sorted)
colors = np.random.rand(num_bars, 3)

# Crear el gráfico de barras
plt.bar(equipos_ordenados.index, equipos_ordenados, color=colors)
plt.xlabel('Equipo')
plt.ylabel('Goles concedidos')
plt.title('Goles concedidos en los minutos 61 al 70 por equipo')

# Rotar las etiquetas del eje x para una mejor legibilidad
plt.xticks(rotation='vertical')

# Mostrar el gráfico de barras
plt.show()
```

```python
# Obtener los datos de goles concedidos en los minutos 61 al 70 por equipo
goles_concedidos = premier.groupby('team_name')['goals_conceded_min_61_to_70'].sum()
goles_concedidos
```

```python
# Ordenar los datos por goles concedidos de forma descendente
equipos_ordenados = goles_concedidos.sort_values(ascending=False)

# Crear el gráfico circular
plt.pie(equipos_ordenados, labels=equipos_ordenados.index, autopct='%1.1f%%', startangle=90, 
       textprops={'fontsize': 5})
      

# Ajustar el aspecto del gráfico
plt.axis('equal')
plt.title('Goles concedidos en los minutos 61 al 70 por equipo')

# Mostrar el gráfico circular
plt.show()
```
