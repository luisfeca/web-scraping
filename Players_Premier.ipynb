*Extraemos las columnas "'full_name', 'Current Club', 'nationality', 'age', 'position'" para su facil interpretacion*

```python
# Seleccionar las columnas 
premier = premier[['full_name', 'Current Club', 'nationality', 'age', 'position']]

# Guardar el nuevo DataFrame en un archivo CSV
premier.to_excel('new premier.xlsx', index=False)
```

```python
premier2=pd.read_excel('new premier.xlsx')
```

```python
premier2
```

```python
# Filtrar por la posición "Midfielder" y que sean del "Bix Six"
Club= ['Manchester City', 'Liverpool', 'Manchester United', 'Arsenal', 'Chelsea', 'Tottenham']
mediocampistas = premier2[(premier2['position'] == 'Midfielder') & (premier2['Current Club'].isin(Club))]
mediocampistas
```

```python
mediocampistas.shape
```

```python
#Calcular el número de mediocampistas por club
BigSixMidfielder = mediocampistas['Current Club'].value_counts()

# Crear el gráfico de barras
plt.bar(BigSixMidfielder.index, BigSixMidfielder, color='green')
plt.xlabel('Club')
plt.ylabel('Cantidad de mediocampistas')
plt.title('Cantidad de mediocampistas por club del "Big Six"')
plt.xticks(rotation=90)

# Mostrar el gráfico
plt.show()
```

```python
# Filtrar por la posición "Forward y que sean <=25
players = premier2[(premier2['position'] == 'Forward') & (premier2['age'] <= 25)]
players
```

```python
# Obtener la cantidad de jugadores "Forward y que sean <=25
club_counts = players['Current Club'].value_counts()

# Crear el gráfico de barras
plt.bar(club_counts.index, club_counts)
plt.xlabel('Club')
plt.ylabel('Cantidad de jugadores jóvenes')
plt.title('Cantidad de jugadores jóvenes en la posición de delantero por club')

# Ajustar el espacio entre las etiquetas del eje x
plt.xticks(rotation=90)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Filtrar jugadores menores o iguales a 25 años en la posición de delantero
players = premier2[(premier2['position'] == 'Forward') & (premier2['age'] <= 27)]
players
```

```python
# Obteniendo los datos de las nacionalidades de los jugadores
nacionalidades = players['nationality'].value_counts()

# Creaamos el gráfico circular
plt.pie(nacionalidades, labels=nacionalidades.index, autopct='%1.1f%%', startangle=90,
        textprops={'fontsize': 9})

# Ajustar el aspecto del gráfico
plt.axis('equal')
plt.title('Distribución de nacionalidades de los jugadores')

# Mostrar el gráfico
plt.show()
```
