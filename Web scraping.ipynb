```python
import pandas as pd

def obtener_total_victorias_totales(df1, df2, df3, equipo):
    total_victorias_totales = 0
    total_victorias_totales += df1.loc[df1['Equipos'] == equipo, 'Victorias Totales'].astype(int).sum()
    total_victorias_totales += df2.loc[df2['Equipos'] == equipo, 'Victorias Totales'].astype(int).sum()
    total_victorias_totales += df3.loc[df3['Equipos'] == equipo, 'Victorias Totales'].astype(int).sum()
    return total_victorias_totales


def pedir_equipo():
    while True:
        equipo = input("Ingresa el nombre del equipo: ")
        if equipo:
            return equipo

while True:
    equipo_deseado = pedir_equipo()
    total_victorias = obtener_total_victorias_totales(df1, df2, df3, equipo_deseado)
    
    if total_victorias > 0:
        print("Total de victorias totales del equipo:", total_victorias)
        break
    else:
        print("El equipo no se encontró en las tablas. Por favor, intenta nuevamente.")
```

```python
import pandas as pd

def obtener_total_derrotas_totales(df1, df2, df3, equipo):
    total_derrotas_totales = 0
    total_derrotas_totales += df1.loc[df1['Equipos'] == equipo, 'Derrotas Totales'].astype(int).sum()
    total_derrotas_totales += df2.loc[df2['Equipos'] == equipo, 'Derrotas Totales'].astype(int).sum()
    total_derrotas_totales += df3.loc[df3['Equipos'] == equipo, 'Derrotas Totales'].astype(int).sum()
    return total_derrotas_totales


def pedir_equipo():
    while True:
        equipo = input("Ingresa el nombre del equipo: ")
        if equipo:
            return equipo

while True:
    equipo_deseado = pedir_equipo()
    total_derrotas = obtener_total_derrotas_totales(df1, df2, df3, equipo_deseado)
    
    if total_derrotas > 0:
        print("Total de derrotas totales del equipo:", total_derrotas)
        break
    else:
        print("El equipo no se encontró en las tablas. Por favor, intenta nuevamente.")
```

```python
url2 = 'https://liquipedia.net/dota2/DreamLeague/Season_20/Statistics'
page2 = requests.get(url2)
soup2 = BeautifulSoup(page2.content,'html.parser')
```

```python
h = soup2.find_all('td',style="text-align:left;white-space:nowrap")
heroes=list()
for i in h:
  heroes.append(i.text)

print(heroes)
```

```python
def remove_nbsp(heroes):
  heroesnew = list()
  for string in heroes:
    heroesnew.append(string.replace("\xa0", " "))
  return heroesnew

heroesnew = remove_nbsp(heroes)
print(heroesnew)
```

```python
longitud = len(heroesnew)
print(longitud)
```

## SACAMOS TODOS LOS DATOS DE LA PAGINA PARA PODER ESCOJER LOS DATOS NECESARIOS Y LIMPIARLOS

```python
t = soup2.find_all('tr')
totale=list()
for i in t:
  totale.append(i.text)

print(totale)
```

## SACAMOS LOS DATOS PARA LA TABLA BASE

```python
import re


new_data = []

for item in totale:
    values = re.findall(r'\d+\.\d*', item)
    if len(values) > 1:
        name = values.pop(0)
        values = [float(value) for value in values]
        new_item = [name] + values
        new_data.append(new_item)

print(new_data)
```

```python
base = pd.DataFrame(new_data)
```

## USAREMOS LOS DATOS DE LA PRIMERA COLUMNA

```python
import re

colum1 = []

for item in new_data:
    new_item = item[0:1]
    colum1.append(new_item)

print(colum1)
```

### En la columna 1 se encuentran los datos de: total de veces jugado el heroe, victorias, derrotas y % Victorias

## SACAMOS LOS DATOS QUE NOS DARAN NUESTRO % DE VICTORIA

```python
import numpy as np
colnew = []

for item in colum1:
    value = float(item[0])
    if value == 100 or str(item).startswith("101") or str(item).startswith("110"):
        colnew.append([0.00])
    elif value == 0 or item[0][1] == "0":
        colnew.append([0.00])
    elif item[0][2] == "0":
        colnew.append([100.00])
    else:
        colnew.append([value % 100])

porcenta = np.array(colnew).flatten().tolist()
porcentajes = [round(elemento, 3) for elemento in porcenta]

porcentajes = porcentajes[:-1]
ultimo_elemento = [-1]
print(porcentajes)



```

```python
lon = len(porcentajes)
print(lon)
```

## TOTAL DE VESCES PICKEADO EL HEROE Y ARREGLAMOS LOS DATOS CON DEFECTO

```python
n1 = []

for item in colum1:
    value = item[0]
    if value == '0.00':
        n1.append([0])
    else:
        n1.append([int(value[:2])])
        
n1 = np.array(n1).flatten().tolist()
n = n1[:-1]
ultimo_elemento = [-1] 

print(n)
```

```python
long = len(n)
print(long)
```

```python
#ARREGLO DE LOS DATOS
for i in range(len(n) - 1):
    if n[i] < n[i+1]:
        n[i+1] = n[i+1] // 10
        
jugados = n

print(jugados)
```

```python
long = len(jugados)
print(long)
```

## CON EL PROCENTAJE DE VICTORIA Y EL TOTAL DE VECES PICKEADO EL HEROE PODEMOS HALLAR LA CANTIDAD DE VICTORIAS DE CADA HEROE

```python
def calcular_resultados(colnew, n):
    win = []  

    for i in range(len(n)):
        valor = round(porcentajes[i] * jugados[i] / 100)  
        win.append(valor)  
    return win

win = calcular_resultados(colnew, n)

 
print(win)
```

```python
longi = len(win)
print(longi)
```

## CON LA CANTIDAD DE VICORIAS Y EL TOTAL DE VECES PICKEADO EL HEROE HALLAMOS LA CANTIDAD DE DERROTAS

```python
def restar_listas(n, win):
    loss = []  # Lista vacía para almacenar los resultados

    for i in range(len(n)):
        diferencia = jugados[i] - win[i]  # Restar el elemento correspondiente de "win" al elemento correspondiente de "n"
        loss.append(diferencia)  # Agregar el resultado a la lista "resultado"

    return loss

loss = restar_listas(n, win)
print(loss)
```

```python
longit = len(loss)
print(longit)
```
